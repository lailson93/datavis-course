```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dados = pd.read_csv('capes_bolsas.csv')

print(dados.head()) # print as 5 primeiras linhas
```

```python
dados.Ano
```

```python
dados['Iniciação Científica'].fillna(0)

bola =(dados['Região'] == 'NORDESTE').sum()


nordeste =( (dados['Região'] == 'NORDESTE') & (dados['Iniciação Científica'] != 0.0 )).sum()
sul = ((dados['Região'] == 'SUL') & (dados['Iniciação Científica'] != 0.0 )).sum()
sudeste = ((dados['Região'] == 'SUDESTE') & (dados['Iniciação Científica'] != 0.0 )).sum()
norte = ((dados['Região'] == 'NORTE') & (dados['Iniciação Científica'] != 0.0 )).sum()
centro_oeste =( (dados['Região'] == 'CENTRO-OESTE') & (dados['Iniciação Científica'] != 0.0 )).sum()

proportions = [norte, nordeste, sul, sudeste, centro_oeste]

# Create a pie chart
plt.pie(
    # using proportions
    proportions,
    
    # with the labels being officer names
    labels = ['norte', 'nordeste', 'sul', 'sudeste', 'centro_oeste'],
    
    # with no shadows
    shadow = True,
    
    # with colors
    colors = ['green','yellow', 'blue', 'red','purple'],
    
#     # with one slide exploded out
#     explode = (0.15 , 0),
    
    # with the start angle at 90%
    startangle = 90,
    
    # with the percent listed as a fraction
    autopct = '%1.1f%%'
    )
# View the plot drop above
plt.axis('equal')

# Set labels
plt.title("Iniciacao Cientifica por Regiao")

# View the plot
plt.tight_layout()
plt.show()
```
